```python
import pandas as pd
import numpy as np

# Set random seed (so output remains same every time)
np.random.seed(42)

# Number of rows (minimum 50)
n = 50

# Generate dataset
data = {
    "Name": [f"Student_{i+1}" for i in range(n)],
    "Roll_No": np.arange(1, n+1),
    "Height_cm": np.random.randint(150, 190, n),
    "Weight_kg": np.random.randint(45, 90, n),
    "Stipend": np.random.randint(5000, 20000, n)
}

df = pd.DataFrame(data)

print("Original Dataset:\n")
print(df)

# Min-Max Normalization formula
# X_normalized = (X - X_min) / (X_max - X_min)

def min_max_normalize(column):
    return (column - column.min()) / (column.max() - column.min())

# Apply normalization to numerical columns
df_normalized = df.copy()

for col in ["Roll_No", "Height_cm", "Weight_kg", "Stipend"]:
    df_normalized[col] = min_max_normalize(df[col])

print("\n\nNormalized Dataset (Min-Max Scaling):\n")
print(df_normalized)
```
