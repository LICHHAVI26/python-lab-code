```python
# -------- CREATE GRAPH DYNAMICALLY --------
def bfs (st,goal,graph):
    
    vis=set()
    queue=[st]
    path=[]
    vis.add(st)
    while queue:
        node=queue.pop(0)
        path.append(node)
        if node==goal:
            return path
        for neighbor in graph[node]:
            if neighbor not in vis:
                vis.add(neighbor)
                queue.append(neighbor)
    return path


# Number of nodes
n = int(input("Enter number of nodes: "))
e = int(input("Enter number of edges: "))
graph = {}

print("Enter node names:")
for i in range(n):
    node = input()
    graph[node] = []


print("Enter edges (u v):")
for i in range(e):
    u = input("First node: ")
    v = input("Second node: ")

    # Add edge (Undirected Graph)
    graph[u].append(v)
    graph[v].append(u)

# Display the created graph
print("\nGraph Representation (Dictionary Form):")
for node in graph:
    print(node, ":", graph[node])
st = input("Enter Initial Node: ")
goal = input("Enter Goal Node: ")

if st not in graph:
    print(f"Start node '{st}' not found in graph!")
elif goal not in graph:
    print(f"Goal node '{goal}' not found in graph!")
else:
    print(f"\nBFS Traversal from '{st}' to '{goal}':")
    result = bfs(st, goal, graph)
    print("Goal Found:", result)
```

```python
def bfs(st, goal, graph):
    vis = set()
    queue = [st]
    vis.add(st)
    order = []

    while queue:
        node = queue.pop(0)
        order.append(node)

        if node == goal:
            print("Order of traversal:", " -> ".join(order))
            return True

        for child in graph[node]:
            if child not in vis:
                vis.add(child)
                queue.append(child)

    print("Order of traversal:", " -> ".join(order))
    return False

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E', 'F'],
    'C': ['A', 'D', 'F'],
    'D': ['B', 'C', 'E', 'F'],
    'E': ['B', 'D', 'F'],
    'F': ['C', 'D', 'E', 'B']
}

print("BFS Traversal:")
result = bfs('A', 'F', graph)
print("Goal Found:", result)
```
