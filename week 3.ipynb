```python
def dfs(st,goal,graph,vis=None):
    if vis is None:
        vis=set()
    print(st)
    if st==goal:
        return True
    vis.add(st)
    for child in graph[st]:
        if child not in vis:
            if dfs(child,goal,graph,vis):
                return True
    return False
graph = {
    'A': ['B','E'],
    'B': ['A','C'],
    'C': ['F','D','B'],
    'D': ['E','C'],
    'E': ['A','D'],
    'F': ['C']
}
dfs('A', 'F',graph)
```
